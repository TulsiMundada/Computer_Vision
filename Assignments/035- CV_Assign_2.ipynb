### Translate the image towards right by 400 and up by 150 pixels.

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = plt.imread('../images/sachin.jpg')
```

```python
plt.title("Original Image")
plt.imshow(img)
plt.axis('off')
```

```python
# Get the dimensions of the image
height, width = img.shape[:2]

# Define the translation matrix
translate_x = 400  # pixels to the right
translate_y = -150  # pixels up
translation_matrix = np.float32([[1, 0, translate_x], [0, 1, translate_y]])

# Perform the translation
translated_img = cv2.warpAffine(img, translation_matrix, (width + translate_x, height + abs(translate_y)))

## Convert BGR image (OpenCV default) to RGB (Matplotlib default)
# translated_img_rgb = cv2.cvtColor(translated_img, cv2.COLOR_BGR2RGB)
```

```python
plt.title("Translated Image")
plt.imshow(translated_img)
plt.axis('off')
plt.show()
```

### Using cv2

```python
image_cv = cv2.imread('../images/sachin.jpg')
```

```python
# Transform using cv2

height = image_cv.shape[0]
width = image_cv.shape[1]

# Define the center point and rotation angle
center = (width//2, height//2)
angle = 60 # Anti clockwise is positive

# Create the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

# Apply the rotation using warpAffine 
rotated_image = cv2.warpAffine(image_cv, rotation_matrix, (height, width))
```

```python
# Show the rotated image
cv2.imshow('Rotated Image',rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Resize the image to 75% of original dimensions.

```python
resized_image = cv2.resize(image_cv, None, fx=0.75, fy=0.75)

cv2.imshow("Original Image", image_cv)
cv2.imshow('Resized Image', resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Flip the image horizontally, diagonally and vertically. Show all four images in the row.

```python
# Flip the image horizontaly
flipped_horizontal = cv2.flip(image_cv,0)

# Flip the image verticaly
flipped_vertical = cv2.flip(image_cv,1)

# Flip the image diagonaly
flipped_diagonal = cv2.flip(image_cv,-1)

# Show the rotated image 
cv2.imshow('Original Image', image_cv)
cv2.imshow('flipped_horizontal', flipped_horizontal)
cv2.imshow('flipped_vertical', flipped_vertical)
cv2.imshow('flipped_diagonal', flipped_diagonal)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Crop the image at, 500,300 to 560, 400 and save in another file

```python
cropped_image = image_cv[500:560, 300:400]

# Show the rotated image 
cv2.imshow('Original Image', image_cv)
cv2.imshow('Cropped Image', cropped_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('cropped_cv.jpg', cropped_image)
```
