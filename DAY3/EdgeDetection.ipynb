```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
img = cv2.imread("../images/lenna.png",0 )
```

```python
## Sobel parameters K ksize , dx , dy, ddepth
sobelx = cv2.Sobel(img,cv2.CV_64F , 1, 0 , ksize=3)  # sobel for horizontal edges
sobely = cv2.Sobel(img,cv2.CV_64F , 0, 1 , ksize=3)  # sobel for vertical edges

### Combine the sobel images 
combined_sobel = cv2.add(sobelx ,sobely)

thresh , binary_sobel = cv2.threshold(combined_sobel , 60 , 255, cv2.THRESH_BINARY)
plt.figure(figsize=(16,8))
plt.subplot(141) , plt.imshow(img, cmap="gray") , plt.title("Original Image")
plt.subplot(142) , plt.imshow(sobelx, cmap="gray") , plt.title("Sobel X")
plt.subplot(143) , plt.imshow(sobely, cmap="gray") , plt.title("Sobel y")
plt.subplot(144) , plt.imshow(combined_sobel, cmap="gray") , plt.title("Final")
plt.show()
```

```python
image = cv2.imread("../images/circle.png" , 0)

sobelx = cv2.Sobel(image,cv2.CV_64F , 1, 0 , ksize=3)  # sobel for horizontal edges
sobely = cv2.Sobel(image,cv2.CV_64F , 0, 1 , ksize=3)  # sobel for vertical edges
combined_sobel = cv2.add(sobelx ,sobely)
thresh , binary_sobel = cv2.threshold(combined_sobel , 50 , 255, cv2.THRESH_BINARY)

edges = cv2.Canny(image,50,150)

plt.figure(figsize=(16,8))
plt.subplot(131) , plt.imshow(image, cmap="gray") , plt.title("Original Image")
plt.subplot(132) , plt.imshow(binary_sobel, cmap="gray") , plt.title("Binary Sobel")
plt.subplot(133) , plt.imshow(edges, cmap="gray") , plt.title("Edges")
plt.show()
```

```python
gray = cv2.imread("../images/circle.png" , 0)
image = cv2.imread("../images/circle.png" )

gray = np.float32(gray)

dst = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)

dst = cv2.dilate(dst, None)

image[dst > 0.01 * dst.max()] = [0,0,255]

cv2.imshow('Harris Corners', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
