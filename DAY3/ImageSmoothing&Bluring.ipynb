```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
img = cv2.imread("../images/lenna.png" )
```

```python
## Define 5 x 5 averaging kernel
kernel = np.ones((5 , 5),np.float32)/25
## Apply the kernel to smooth the image
smoothed_image = cv2.filter2D(img,-1,kernel)

## Show original & smoothed images
cv2.imshow("Original" , img)
cv2.imshow("Smoothed" , smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
new_image = np.hstack([img, smoothed_image])
cv2.imshow("Both Images" , new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Gaussian Kernel

```python
kernel_size = 7   # Adjust as needed
sigma = 0.35 * ((kernel_size - 1 )* 0.5 -1) + 0.9  # Gaussian Standard Deviatian
kernel = cv2.getGaussianKernel(kernel_size , sigma)
```

```python
sigma
```

```python
kernel
```

```python
smoothed_image = cv2.filter2D(img,-1,kernel)

## Show original & smoothed images
new_image = np.hstack([img, smoothed_image])
cv2.imshow("Both Images" , new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Median Filter

```python
gray = cv2.imread('../images/lenna.png' ,0)
```

```python
gray.shape
```

```python
kernel_size = 15
median = cv2.medianBlur(gray , kernel_size)
plt.figure(figsize=(10,6))
plt.subplot(121) , plt.imshow(gray, cmap="gray") , plt.title("Original Image")
plt.subplot(122) , plt.imshow(median, cmap="gray") , plt.title("Median Filtered Image")
plt.show()
```

```python
gray = cv2.imread('../images/lenna.png')
blurred_patch = gray[100:300, 100:300]
blurred_patch = cv2.medianBlur(blurred_patch , ksize=25)  
gray[100:300, 100:300] = blurred_patch
plt.figure(figsize=(10,6))
plt.subplot(121) , plt.imshow(gray, cmap="gray") , plt.title("Original Image")
plt.show()
```
